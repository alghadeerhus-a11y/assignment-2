---
## Part 2: Machine Learning Problem Definition

**Problem Type:** Binary Classification  
**Target Variable:** `Outcome` (0 = No Diabetes, 1 = Diabetes)  

**Problem Statement:**  
Given medical diagnostic measurements (glucose level, BMI, age, blood pressure, etc.) for female patients, predict whether a patient has diabetes or not.

**Why is this Classification (not Regression or Clustering)?**  
- The output (`Outcome`) is a **discrete categorical label** (0 or 1), not a continuous number → so it is NOT regression  
- We have **labeled data** (each row already has a known outcome) → so it is NOT clustering  
- We are predicting one of two categories → this is **binary classification**

## Part 3: Loading and Inspecting the Dataset

```python
# Import required library
import pandas as pd
```

```python
# Load the dataset
# Download from: https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

column_names = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'
]

df = pd.read_csv(url, names=column_names)
print("Dataset loaded successfully!")
```

```python
# 1. Display the shape (rows, columns)
print("Dataset Shape:")
print(f"  Rows (patients): {df.shape[0]}")
print(f"  Columns (features): {df.shape[1]}")
```

```python
# 2. Preview the first 5 rows
print("First 5 rows of the dataset:")
df.head()
```

```python
# 3. Check column names and data types
print("Column names and data types:")
print(df.dtypes)
```

```python
# 4. Basic dataset information
print("Dataset Info:")
df.info()
```

```python
# 5. Check target variable distribution
print("Target Variable Distribution (Outcome):")
print(df['Outcome'].value_counts())
print(f"\n0 = No Diabetes: {df['Outcome'].value_counts()[0]} patients")
print(f"1 = Diabetes:     {df['Outcome'].value_counts()[1]} patients")
```
